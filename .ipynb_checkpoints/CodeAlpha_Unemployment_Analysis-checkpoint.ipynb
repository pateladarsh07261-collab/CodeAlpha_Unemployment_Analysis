```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
```

```python
df.head()
```

```python
print(df.columns.tolist())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip()
```

```python
df.isnull().sum()
```

```python
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
```

```python
state = df.groupby("Region")["Estimated Unemployment Rate (%)"].mean()

state.sort_values(ascending=False).head(10)
```

```python
top10 = state.sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top10.index, top10.values)
plt.xticks(rotation=45)
plt.title("Top 10 States by Unemployment Rate")
plt.xlabel("State")
plt.ylabel("Unemployment Rate")
plt.show()
```

```python
monthly = df.groupby("Date")["Estimated Unemployment Rate (%)"].mean()

plt.figure(figsize=(10,5))
plt.plot(monthly.index, monthly.values, marker="o")
plt.title("Monthly Unemployment Trend")
plt.xlabel("Date")
plt.ylabel("Rate")
plt.show()
```

```python
employment = df.groupby("Region")["Estimated Employed"].mean()

top10_emp = employment.sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top10_emp.index, top10_emp.values)
plt.title("Top 10 States by Estimated Employment")
plt.xlabel("States")
plt.ylabel("Estimated Employed")
plt.xticks(rotation=45)
plt.show()
```

# Conclusion

• Unemployment increased during the COVID-19 period.

• Some states recorded much higher unemployment than others.

• Employment levels varied significantly across regions.

• Data visualization helps understand unemployment trends and regional differences.
